```python
import pandas as pd
df1 = pd.DataFrame(data=content,
             index=[f"第{i}號"for i in range(1,51)],
             columns=['國文','英文','數學','地理','歷史'])
df1
```

```python
df1.loc["第1號",:]
```

```python
df1.loc[["第1號","第5號","第10號"]]
```

```python
student_sum = df1.sum(axis=1)
student_mean = df1.mean(axis=1)
df1['總分'] = student_sum
df1['平均'] = student_mean
df1
```

```python
df1.loc['第10號':'第15號']
```

```python
df2=df1.sort_values(by='總分',ascending=False)
df2
```

```python
df2.iloc[0:5]
```

```python
df2.iloc[-5:]
```

```python
df2['排名'] = df1['總分'].rank(method='first',ascending=False)
df2
```
